## Ejemplo de Regresión Lineal General

El objetivo de este ejemplo es determinar los efectos de los abusos sexuales en mujeres adultas.  En la base de datos hay 45 mujeres tratadas que sufrieros abusos sexuales en la    infancia. Y 31 mujeres que están siendo tratadas en el mismo centro pero que no sufrieron abusos en su infancia.

```r
# cargamos los datos sexab de la libreria faraway
library(faraway)
data(sexab)
# vemos los tipos de variables que hay en la base de datos
str(sexab)
```

Las variables son las siguientes:
* csa: abusos sexuales en la infancia (children sexual abuse): abused / not abused
* ptsd: estres post traumático (postraumatic stress disease): numerica
* cpa: abusos físicos en la infancia (children physical abuse): numerica

Vemos los datos...

```r
sexab
```

```r
# veamos el numero de abusadas y no abusadas
summary(sexab$csa)
```

```r
# resumen de los estadísticos según si han sufrido abuso o no
by(sexab,sexab$csa,summary) 
```

Parece que el estres postraumatico es mayor en las abusadas. Vamos a dibujar los datos según si han sido abusadas o no

```r
plot(ptsd~csa, sexab)
plot(ptsd~cpa, pch=as.character(csa),sexab)
```

```r
#Hacemos un test para ver si esas diferencias son estadísticamente significativas:
t.test(sexab$ptsd[1:45],sexab$ptsd[46:76])  
## Se rechaza la hipótesis nula de igualdad de medias
```

Hacemos el modelo lineal general 

$$
Y = \beta_0 + \beta_1 x_1 + \beta_1 x_2
$$

donde 
* $x_1$: cpa
* $x_2$: csa


```r
ej3_sin_interaccion<-lm(ptsd ~ cpa + csa, data = sexab)

# vemos la matrix X del modelo
model.matrix(ej3_sin_interaccion)
```

Vemos que las abusadas han sido codificadas como 0 y las no abusadas como 1. Por tanto, se ha tomado como referencia las abusadas. Veamos la significación de los parámetros.

```r
summary(ej3_sin_interaccion)
```

Vemos que ambos parámetros son significativos y que el p-valor del contraste F también es pequeño. Por tanto, hay confusión, es decir, para las abusadas tenemos
$$
Y= 10.2480 + 0.5506 * x_1  
$$
mientras que para las no abusadas
$$
Y= 10.2480 + 0.5506 * x_1 -6.2728
$$

Hacemos el modelo lineal general considerando la interacción

$$
Y = \beta_0 + \beta_1 x_1 + \beta_1 x_2 + \beta_3 x_1 x_2
$$


```r
ej3_con_interaccion<-lm(ptsd ~ cpa + csa + cpa*csa, data = sexab)

# vemos la matrix X del modelo
model.matrix(ej3_con_interaccion)

# veamos la significación de los parámetros
summary(ej3_con_interaccion)
```

Vemos que no hay interacción ya que el parámetro de la interacción ( $ x_1 * x_2 $ ) no es significativo. Además, vemos que el modelo lineal sin interacción tiene todos los parámetros significativos, por tanto no podemos reducirlo más.

Vamos a dibujar las rectas de regresión

```r
plot(ptsd~cpa, pch=as.character(csa),sexab)
abline(10.5571,0.45)
abline(10.5571-6.8612,0.45,lty=2)
```

### Diagnosis del modelo

```r
g<- ej3_sin_interaccion
plot(fitted(g),residuals(g),pch=as.character(sexab$csa),xlab="Fitted", ylab="Residuals")
```

Aceptamos que la varianza de los residuos es constante para las abusadas y para las no abusadas

```r
res.estan.g<-residuals(g)/(summary(g)$sig*sqrt(1-influence(g)$hat)) 
ks.test(res.estan.g,pnorm,mean(res.estan.g),sd(res.estan.g))
```

Aceptamos la normalidad de los residuos
